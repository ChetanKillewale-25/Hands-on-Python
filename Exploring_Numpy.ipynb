# Some Basic Operations

```python
a = np.array([range(i,i+3) for i in [2,4,6]])
print(a)
#   A 3*5 floating array filled with zeros : 
a = np.zeros((3,5),dtype=int)
print(a)
#   length-10 integer array filled with zero : 
a = np.zeros(10,dtype=float)
print(a)
#   Array filled with 3.14 : 
a = np.full((3,2),3.14)
print(a)
#   Starting at 0 ending at 20, stepping by 2 : 
a = np.arange(0,20,2)
print(a)
#   An array of five values evenly spaced between 0 & 1 :
a = np.random.random((3,3))
print(a)
#   Identity Matrix:
a = np.eye(3)
print(a)
#   Elements from 0 to 9 in a very shuffled
a = np.random.permutation(np.arange(10))
print(a)
#   Returning values between 20 and 30
a = np.random.randint(20,30)
print(a)
print(type(a))

a = np.random.rand(1000)
print(a)   #Random numbers from 0 to 1

a = np.random.rand(2,3,4,2)
print(a)
"""It contains two arrays that are 3-dimensional, each dimensional array has 3-dimensional,
 each dimensional array has 4-dimensional and one dimensional array has 2 elements in it.
"""

a = np.arange(100).reshape(4,25)
print(a)
a = np.arange(100).reshape(4,5,5)
print(a)

"""
ARRANGING, RESHAPING & SHAPING THE ARRAYS.
"""
x1 = np.random.randint(10,size=6)
print(x1)
```

# UFUNC'S: Remove slow Python Loops

```python
# Absolute Value:
x = np.array([[-2,-1,0,1,2]])
print(np.abs(x))
# Absolute also returns the magnitude of Imaginary Numbers:
x = np.array([3-4j,4-3j,2+0j])
print(np.absolute(x))

#   Triggnometric Function
theta = np.linspace(0,np.pi,3)
print(theta)
print(np.sin(theta))

#   Inverse Trignomrteric Function:
x = [-1,0,1]
print(x)
print(np.arcsin(x))
print(np.arccos(x))

#   Exponents and Logarithms
x = [1,2,3]
print(x)
print(np.exp(x))
print(np.exp2(x))
print(np.power(3,x))

x = [1,2,4,10]
print(np.log(x))
print(np.log2(x))
print(np.log10(x))

"""
When x is very small, thes func. gives more precise values.....
"""
x = [0,0.001,0.01,0.1]
print(np.expm1(x))
print(np.log1p(x))

# Some special Ufuncs: 

from scipy import special
#   Gamma Function: 
x = [1,5,10]
print(special.gamma(x))
print(special.gammaln(x))
print(special.beta(x,2))

#   Error  Function (Integral of Gaussain) : it's compliment and it's inverse....
x = np.array([0,0.3,0.7,1.0])
print(special.erf(x))
print(special.erfc(x))
print(special.erfinv(x))

#   Specifying the Output:
x = np.arange(5)
y = np.empty(5)
np.multiply(x,10,out=y)
print(y)
y = np.zeros(10)
np.power(2,x,out=y[::2])
print(y)

#   Aggreagates: 
x = np.arange(1,6)
print(np.add.reduce(x))
print(np.multiply.reduce(x))
print(np.add.accumulate(x))
print(np.multiply.accumulate(x))


# Outer Products: 
x = np.arange(1,6)
print(np.multiply.outer(x,x))
```

# BROADCASTING: Simply a set of rules on applying Binary Ufuncs to on arrays of different sizes......

```python
A = np.round(10*np.random.rand(2,3))
print(A+3)
print(A+(np.arange(2).reshape(2,1)))

B = np.round(10*np.random.rand(2,2))
print(B)
C = np.hstack((A,B))
print(C)

#   Splittting of Arrays: 
grid = np.arange(16).reshape((4,4))
print(grid)
uppper,lower = np.vsplit(grid,[2])
left,rightz  = np.hsplit(grid,[2])

#   Comparison Operator as Ufuncs:
"""
Numpy also implements comparison operators such as <(less than) and >(grater than) 
as element-wise Ufuns. The result of the copmoarison operator is always an array with a boolean
data type.....
"""
x = np.array([1,2,3,4,5])
print(x<3)
print(x>3)
print(x<=3)
print(x>=3)
print(x!=3)
print(x==3)

"""
It is also possible to do an element-by-element comparison of two arrays
and to include compound expressions...
"""
print((2*x) == (x**2))
```

# Working with Boolean Arrays:

```python
S = np.array([[5,0,3,3],
              [7,9,3,5],
              [2,4,7,6]
])

print(np.count_nonzero(S<6))
print(np.sum(S<6))
print(np.sum(S<6,axis=1))   # This counts the number of values less than 6 in each row of the matrix..

#   Any and All
#      Are there any values greater than 8
print(np.any(S>8))
print(np.any(S==1))
#      Are all values less than 0
print(np.all(S<0))
print(np.all(S==7))

#   Boolean Array as Masks:  
print(S<5)
# ...Now to select this values from the array:  
S[S<5]
```
